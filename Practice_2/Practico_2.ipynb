Normalmente hay un exceso de carga negativa en la superficie interior de la membrana celular de una neurona y una carga positiva de equilibrio en su superficie exterior.

En esta disposición, la membrana celular crea una capacitancia $C_m$, y el voltaje a través de la membrana $V $ y la cantidad de este exceso de carga $ Q $ están relacionados por la ecuación estándar para un capacitor, $Q = C_m V $.

La capacitancia de la membrana es proporcional a la cantidad total de membrana o, de manera equivalente, al área de superficie de la celda. La capacitancia total $C_m$ es el área de superficie de la membrana $A$ multiplicada por la capacitancia específica, $ C_m = c_m A $.

Podemos usar la capacitancia de la membrana para determinar cuánta corriente se requiere para cambiar el potencial de la membrana a una velocidad determinada. La derivada en el tiempo de la ecuación básica que relaciona el potencial de membrana y la carga,

$$C_m\dfrac{dV}{dt} = \dfrac{dQ}{dt}$$

juega un papel importante en el modelado matemático de neuronas. La 
derivada de la carga con respecto al tiempo $\dfrac{dQ}{dt}$ es igual a la corriente que pasa al
célula, por lo que la cantidad de corriente necesaria para cambiar el potencial de membrana
de una neurona con una capacitancia total $ C_m $ a una tasa $\dfrac{dV} {dt} $ es $ C_m \dfrac {dV} {dt} $.

La capacitancia de una neurona determina cuánta corriente se requiere para
hacer que el potencial de membrana cambie a una velocidad determinada. Mantener el potencial de membrana estable a un nivel diferente de su valor de reposo también requiere corriente, pero esta corriente está determinada por la resistencia de la membrana en lugar de por la capacitancia de la célula.


$R_m $ se conoce como resistencia de entrada o de membrana. Se requiere la restricción a pequeñas corrientes y pequeños $V$ porque las resistencias de las membranas pueden variar en función del voltaje, mientras que la ley de Ohm asume que $R_m$ es constante en el rango $ V $.

La resistencia de la membrana es la inversa de la conductancia de la membrana,
y, como la capacitancia, la conductancia de un trozo de membrana celular
es proporcional a su superficie. La constante de proporcionalidad es la
conductancia de la membrana por unidad de área, pero la escribimos como $ \dfrac {1} {r_m} $, donde $ r_m $ es
llamada resistencia específica de la membrana. Por el contrario, la resistencia de la membrana $ R_m $ es igual a $r_m$ dividido por el área de la superficie.

El producto de la capacitancia de la membrana y la resistencia de la membrana es
una cantidad con las unidades de tiempo llamada constante de **tiempo de membrana**, $ \tau_m = R_m C_m $. Dado que $ C_m $ y $ R_m $ tienen dependencias inversas del área de superficie de la membrana, la constante de tiempo de la membrana es independiente del área e igual
al producto de la capacitancia y la resistencia específicas de la membrana, $ \tau_m = r_m c_m $.

La corriente total que fluye a través de la membrana a través de todos sus canales iónicos se denomina corriente de membrana de la neurona. Por convención, la corriente de membrana se define como positiva cuando los iones positivos salen de la neurona y negativa cuando los iones positivos entran en la neurona. La corriente total de la membrana se determina sumando las corrientes debidas a todos los diferentes tipos de canales dentro de la membrana celular, incluidos los dependientes del voltaje y canales sinápticos. Para facilitar las comparaciones entre neuronas de diferentes tamaños, es conveniente utilizar la corriente de membrana por unidad de área de membrana celular, que llamamos $i_m$. La corriente total de membrana se obtiene de $i_m$ multiplicándolo por $A$, el área de superficie total de la celda.

Marcamos los diferentes tipos de canales en una membrana celular con un índice $ i $. La corriente transportada por un conjunto de canales de tipo $ i $ con potencial de inversión $ E_i $, desaparece cuando el potencial de membrana satisface $V =E_i$. Para muchos tipos de canales, la corriente aumenta
o disminuye aproximadamente linealmente cuando el potencial de membrana se desvía de este valor. La diferencia $V - E_i$ se llama fuerza motriz, y la corriente de membrana por unidad de área debida a los canales de tipo $ i $ se escribe
como $ g_i (V - E_i) $ El factor $ g_i $ es la conductancia por unidad de área, o
conductancia, debido a estos canales. Sumando los diferentes tipos de canales, obtenemos la corriente total de membrana.

$$i_m = \sum_i g_i(V-E_i)$$

Gran parte de la complejidad y riqueza de la dinámica neuronal surge porque
las conductancias de la membrana cambian con el tiempo. Sin embargo, algunos de los factores que contribuyen a la corriente total de la membrana pueden tratarse como relativamente constantes y, por lo general, se agrupan en un solo término llamado *corriente de fuga*. Las corrientes transportadas por bombas de iones que mantienen
mantener los gradientes de concentración que hacen que los potenciales de equilibrio sean distintos de cero
normalmente entran en esta categoría.

Todo las contribuciones independiente de tiempo a la corriente de la membrana se pueden agrupar en un solo término de fuga $ \overline {g} _L(V - E_L) $. La línea sobre el parámetro $ \overline {g} _L $ se usa para indicar que tiene convalor constante.

#### Modelos monocomportamentales

Los modelos que describen el potencial de membrana de una neurona mediante una única variable $ V $ se denominan modelos de un solo compartimento.

La ecuación

$$C_m\dfrac{dV}{dt} = \dfrac{dQ}{dt}$$ 

proporciona la relación básica que determina el potencial de membrana para un modelo de un solo compartimento. Esta ecuación establece que la tasa de cambio del potencial de membrana es proporcional a la tasa
en el cual la carga se acumula dentro de la célula. La tasa de acumulación de carga es,
a su vez, igual a la cantidad total de corriente que ingresa a la neurona. Las corrientes relevantes son las que surgen de toda las conductancias membrana y sináptica más, en un entorno experimental, cualquier corriente inyectada en la celda a través de un electrodo.

De la ecuación, la suma de estas corrientes es igual a $C_m \dfrac{dV}{dt}$, la capacitancia total de la neurona multiplicada por la tasa de cambio del potencial de membrana. Debido a que la corriente de membrana generalmente se caracteriza como una corriente por unidad de área, $ i_m $, es más conveniente dividir esta relación por el área de superficie de la neurona. Entonces, el total
la corriente por unidad de área es igual a $ c_m \dfrac{dV}{dt}$, donde $c_m = \dfrac{C_m}{A} $ es la capacitancia específica de la membrana.

Una complicación de este procedimiento es que la corriente del electrodo, $I_e$, no se expresa típicamente como una corriente por unidad de área, por lo que debemos dividirla por el área de superficie total de la neurona, $A$.

Juntando todo esto, la ecuación básica para todos los modelos de un solo compartimento es

$$c_m \dfrac{dV}{dt} = -i_m + \dfrac{I_e}{A}$$

Por convención, la corriente que ingresa a la neurona a través de un electrodo se define como positiva hacia adentro, mientras que la corriente de membrana se define como positiva hacia afuera.

#### Modelos Integrate-and-Fire

Una neurona normalmente disparará un potencial de acción cuando su potencial de membrana alcance un valor umbral de aproximadamente -55 a -50 mV. Durante el potencial de acción, el potencial de membrana sigue una trayectoria rápida y estereotipada y luego vuelve a un valor que está hiperpolarizado en relación con el potencial umbral. Como veremos, los mecanismos mediante los cuales las conductancias $ K ^ {+} $ y $ Na ^ {+} $ dependientes del voltaje producen potenciales de acción se comprenden bien y se pueden modelar con bastante precisión.

Por otro lado, los modelos de neuronas pueden simplificarse y las simulaciones pueden acelerarse drásticamente si los mecanismos biofísicos responsables de los potenciales de acción no se incluyen explícitamente en el modelo.

Los modelos de integración y disparo hacen esto estipulando que un potencial de acción ocurre siempre que el potencial de membrana de la neurona modelo alcanza un valor umbral $ V_ {th} $. Después del potencial de acción, el potencial se restablece a un valor $ V_ {reset} $ por debajo del potencial umbral, $ V_ {reset} < V_ {th} $.

Lapicque propuso el modelo básico de integración y disparo en 1907, mucho antes de que se entendieran los mecanismos que generan los potenciales de acción. A pesar de su antigüedad y simplicidad, el modelo de integración y activación sigue siendo una descripción extremadamente útil de la actividad neuronal.

Al evitar una descripción biofísica del potencial de acción, se dejan los modelos de integración y disparo.
con la tarea más simple de modelar solo la dinámica del potencial de membrana por debajo del umbral. Esto se puede hacer con varios niveles de rigor. En la versión más simple de estos modelos, se ignoran todas las conductancias de membrana activas,incluyendo, por el momento, entradas sinápticas, y toda la conductancia de la membrana se modela como un solo término de fuga pasiva, $ i_m = \overline {g}_L (V - E_L)$.

Para pequeñas fluctuaciones sobre el potencial de membrana en reposo, las conductancias neuronales son aproximadamente constantes, y el modelo pasivo de integración y disparo supone que esta constancia se mantiene en todo el rango de subumbrales. Para algunas neuronas, esta es una aproximación razonable y para otras no lo es. Con estas aproximaciones, la neurona modelo se comporta como un circuito eléctrico que consta de una resistencia y un capacitor en paralelo y el potencial de membrana es

$$c_m \dfrac{dV}{dt} = -\overline{g}_L (V − E_L) + \dfrac{I_e}{A}$$

Es conveniente multiplicar esta ecuación por la resistencia específica de la membrana $ r_m $, que en este caso viene dada por $ r_m = \dfrac {1} {\overline {g} _L} $. Esto cancela el factor de $ \overline {g} _L $ en el lado derecho de la ecuación y deja un factor $ c_m r_m = \tau_m $ a la izquierda
lado, donde $ \tau_m $ es la constante de tiempo de membrana de la neurona. El electrodo
la corriente termina siendo multiplicada por $ \dfrac {r_m} {A} $, que es la resistencia total de la membrana $ R_m $. Por tanto, la ecuación básica de los modelos pasivos de integración y disparo es
$$c_m \dfrac{dV}{dt} = E_L - V  + I_e R_m$$

Para generar potenciales de acción en el modelo, la última ecuación es aumentada por
la regla de que siempre que $ V $ alcanza el valor umbral $V_{th}$, se dispara un potencial de acción y el potencial se restablece a $V_{reset}$. Esta cita indica que cuando $I_e = 0$, el potencial de membrana se relaja exponencialmente con la constante de tiempo $ \tau_m $ a $V = E_L$. Por tanto, $ E_L $ es el potencial de reposo de la célula modelo.

El potencial de membrana para el modelo pasivo de integración y fuego se determina integrando la ecuación y aplicando la regla de umbral y restablecimiento para la generación de potencial de acción. Se muestra la respuesta de una neurona modelo pasiva de integración y disparo a una corriente de electrodo variable en el tiempo.

**Problema 1**

Considerá el modelo *Integrate-and-Fire* para la evolución temporal del potencial de membrana $V_m(t)$ al tiempo $t$ entre el interior y el exterior de una neurona genérica.

$$\tau_m \dfrac{\partial V_m(t)}{\partial t} = E_L + R_m I_e(t) - V_m(t)$$

donde $E_L$ es el potencial en reposo, $I_e(t)$ es una corriente eléctrica externa (cuyo valor positivo corresponde
a una corriente entrante) que se inyecta (input), $R_m$ es la resistencia y $\tau_m$ es el tiempo característico
de la membrana $\tau_m = r_m c_m$ (donde $r_m$ y $c_m$ son respectivamente la resistencia y la capacitancia de la
membrana por unidad de área). Esta ecuación se puede reescribir como:

$$ \dfrac{d V_m(t)}{d t} = \dfrac{1}{\tau_m}(E_L + R_mI_e(t) - V_m(t))$$

a) Resolvé analíticamente esta ecuación sin incorporar el umbral de disparo para el caso de una corriente
externa constante $I_e(t) = I_e$ y $V_m(t=0) = V_0$. Grafica la solución para $0 ms ≤ t ≤ 200 ms$ con los
siguientes valores de los parámettros:

$$Vm(t = 0) = E_L = −65 mV,\hspace{1em} R_m = 10 M\Omega, \hspace{1em} Vth = −50 mV, \hspace{1em} \tau_m = 10 ms$$
Discutí e interpretá

<font color='green'> RESPUESTA </font>

Si $I_e(t) = I_e$, es decir, una constante e independiente del tiempo entonces podemos reescribir la ecuación anterior como:

$$\dfrac{d V_m(t)}{d t} = \dfrac{1}{\tau_m}(E_L + R_m I_e - V_m(t))$$

Sea $V_{\infty} = E_L + R_m I_e$ (una constante) y definamos una nueva variable $Z(t) = V_{\infty} - V_m(t)$, entonces:  

$$\dfrac{d V_m(t)}{d t} = \dfrac{V_{\infty} - V_m(t)}{\tau_m} = \dfrac{Z(t)}{\tau_m} \hspace{3em} $$

Por otro lado la derivada de $Z(t)$ con respecto a $t$:

$$\dfrac{dZ(t)}{dt} = \dfrac{dV_{\infty}}{dt} - \dfrac{d V_m(t)}{dt} = -\dfrac{d V_m(t)}{dt} = - \dfrac{Z(t)}{\tau_m}$$

Entonces podemos reescribir la ecuacion anterior como:

$$\dfrac{dZ(t)}{Z(t)} = -\dfrac{dt}{\tau_m}$$

e integrando ambos lados como:

$$\int_{Z(0)}^{Z(t)} \frac{dZ'(t)}{Z'(t)} = \ln{\frac{Z(t)}{Z(0)}} = -\frac{1}{\tau_m}\int_{0}^{t} dt' = -\frac{t}{\tau_m}$$

Esto nos da 

$$Z(t) = Z(0)exp\left(− \dfrac{t}{\tau_m}\right) \iff V_{\infty} - V_m(t) = (V_{\infty} - V_m(0))exp\left(− \dfrac{t}{\tau_m}\right)$$

Si $V_m(t=0) = V_0$ entonces

$$V_m(t) = V_{\infty} - (V_{\infty} - V_0)exp\left(− \dfrac{t}{\tau_m}\right)$$

$$\iff V_m(t) = E_L + R_m I_e + (V_0 - E_L - R_m I_e)exp\left(− \dfrac{t}{\tau_m}\right)$$

```python
import numpy as np

# seteamos los parámetros
params = {'I_e': 2, 
          'E_L': -65,
          'V_0': -65,
          'R_m': 10,
          'Vth': -50,
          'tau_m': 10}

def V_m(t, params):
    """
    Retorna el valor del potencial de mebrana en el tiempo t
    t: tiempo en segundos (s).
    parms: conjuntos de parámetros para el modelo integrate-and-fire.
    """
    I_e = params['I_e']
    E_L, V_0 = params['E_L'], params['V_0']
    R_m = params['R_m']
    Vth = params['Vth']
    tau_m = params['tau_m']
    
    Z_0 = V_0 - E_L - R_m*I_e
    return E_L + R_m*I_e + Z_0*np.exp(-t/tau_m)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

## Parámetros para graficar
sns.set_style('whitegrid')
sns.set_context('paper')


ts = np.arange(0, 200, 0.1)
vs = V_m(ts, params)

fig = plt.figure(figsize=(10, 4))
plt.plot(ts, vs, color='red')
plt.title('Potencial de membrana sin umbral')
plt.xlabel('Tiempo (ms)')
plt.ylabel('Voltaje (mV)')
#plt.ylim(-65.5, -47);
plt.ylim(-65)
plt.xlim(-0.4, 200.5)
plt.tight_layout()
plt.savefig('./images/sin_umbral.png', dpi=400);
```

b) Usá el método de Runge Kutta de cuarto orden para resolver el problema de valor inicial del modelo
Integrate-and-Fire:

$$\dfrac{d V_m(t)}{d t} = \dfrac{1}{\tau_m}(E_L - V_m(t)+ R_m I_e(t))\hspace{2em} con \hspace{0.5em} V(t = 0) = E_L,\hspace{0.5em} 0 ms ≤ t ≤ 200 ms,\hspace{0.5em} y \hspace{0.5em} h = 0.05 s$$
donde $h$ es el paso de integración y los parámetros toman los valores usados en el punto a). Tené presente
que ahora debés agregar en la simulación el umbral de disparo propio del modelo *Integrate-and-Fire*.
O sea, si $V_m(t)$ ultrapasa el valor umbral $V_{th}$, debés restituir el valor de $V_m(t)$ a $E_L$. 

La corriente externa $I_e(t)$ debe ser constante y tomar el valor $I_e = 2 nA$. Graficá la aproximación numérica de $V_m(t)$ superpuesta con la solución analítica del punto a) (que no tenía disparos) de $Vm(t)$ en $0 ms ≤ t ≤ 200 ms$.

```python
from scipy.integrate import solve_ivp

## Parámetros de integración
h  = 0.05                          ## Paso temporal 
t0 = 0                            ## Tiempo inicial
tf = 200                           ## Tiempo final

## Instantes donde evaluar la solución.
t_eval = np.arange(t0, tf, h)


def dVm_dt(v, I_e=params['I_e']):
    E_L = params['E_L']
    R_m = params['R_m']
    tau_m = params['tau_m']
    return (E_L - v + R_m*I_e)/tau_m

def Runge_Kutta(fun, y_n, h, I_e):
    k1 = fun(y_n, I_e)*h
    k2 = fun(y_n + 0.5*k1, I_e)*h
    k3 = fun(y_n + 0.5*k2, I_e)*h
    k4 = fun(y_n + k3, I_e)*h
    y_n1 = y_n + ((k1 + 2*k2 + 2*k3 + k4)/6) 
    return y_n1

## Condición inicial
def disparo_con_umbral(Ie):    
    ## Parámetros de integración
    h  = 0.05                          ## Paso temporal 
    t0 = 0                            ## Tiempo inicial
    tf = 200                           ## Tiempo final

    ## Instantes donde evaluar la solución.
    t_eval = np.arange(t0, tf, h)

    V0 = params['E_L'] 
    V = V0
    vs_RK = []
    for t in t_eval:
        V = Runge_Kutta(dVm_dt, V, h, Ie)
        if V > params['Vth']:
            V = V0
        vs_RK.append(V)
    return vs_RK

vs_RK = disparo_con_umbral(params['I_e'])
```

```python
fig = plt.figure(figsize=(12, 4))
plt.plot(ts, vs, label='Sol. analítica sin disparos', color='red')
plt.plot(t_eval, vs_RK, label='Runge Kutta. con disparos', color='green')
plt.suptitle('Potencial de membrana con y sin disparos')
plt.xlabel('Tiempo (ms)')
plt.ylabel('Voltaje (mV)')
#plt.ylim(-65.5, -47);
plt.ylim(-65)
plt.xlim(-0.4, 200.5)
plt.legend(bbox_to_anchor=(0.77, 0.77))
plt.savefig('./images/con_umbral.png', dpi=200);
```

c) Ahora varía los valores de $I_e$ entre $0$ y $6$ y calculá para cada valor la frecuencia de disparo. Grafica
la frecuencia de disparo $\omega$ vs. $I_e$ (recordá la relación entre frecuencia y período). Intentá resolver esta
ecuación $\omega(I_e)$ analíticamente (no es obligatorio esto).

```python
I_es = np.arange(0, 7, 1)

## Parámetros de integración
h  = 0.1                          ## Paso temporal 
t0 = 0                              ## Tiempo inicial
tf = 1000                           ## Tiempo final (1000 ms = 1 seg)

## Instantes donde evaluar la solución.
t_eval = np.arange(t0, tf, h)

freqs = []

## Condición inicial
V0 = params['E_L'] 
V = V0
vs_RK = []
for I_e in I_es:
    count = 0
    for t in t_eval:
        V = Runge_Kutta(dVm_dt, V, h, I_e)
        if V > params['Vth']:
            V = V0
            count += 1
    freqs.append(count)
```

```python
periods = []
for freq in freqs:
    if freq != 0:
        periods.append((1/freq)*1e3)
    else:
        periods.append(freq)
```

```python
fig = plt.figure(figsize=(12, 5))
plt.plot(I_es, periods, label='Simulaciones de periodos')
plt.legend();
```

```python
def periodo_analitico(I_e):
    if I_e <= 1.5:
        return 0
    else:
        return 10*np.log((-10*I_e)/(15 - 10*I_e))

I_eval = np.arange(0, 6.01, 0.01)
per_analitic = [periodo_analitico(x) for x in I_eval]
fig = plt.figure(figsize=(10, 5))
plt.plot(I_es, periods, 'o', label='Simulaciones', color='b')
plt.plot(I_eval, per_analitic, label='Sol. analítica', color='red')
plt.xlabel('Corriente externa (nA)', fontsize=12)
plt.ylabel('Periodo (ms)')
plt.legend()
plt.suptitle('Periodo vs Corriente externa')
plt.savefig('./images/periodo_corriente.png', dpi=200);
plt.tight_layout();
```
